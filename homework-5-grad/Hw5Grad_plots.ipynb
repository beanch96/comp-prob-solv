```python
import Psi_1s_func as fn 
import matplotlib.pyplot as plt
import numpy as np

#Diagonal Kinetic Energy Matrix Element (Kii)-Random sampling

# Parameters
np.random.seed(42)
L = 7  # Integration limits
n_points_list = [10**2, 10**3, 10**4, 10**5, 10**6, 10**7,10**8]  # Varying number of points
R=0  #No separation distance-diagonal elements

# Lists to store results
Kii_values = []
std_devs = []

# Monte Carlo integration loop for varying n_points
for n_points in n_points_list:
    Kii,std_dev=fn.random_integration(L,n_points,R)
    Kii_values.append(Kii)
    std_devs.append(std_dev)
    print(f"N = {n_points}, K_ii = {Kii}, Std Dev = {std_dev}")
   
# Plot the results
plt.figure(figsize=(8, 6))
plt.errorbar(n_points_list, Kii_values, yerr=std_devs, fmt='o-', label='Estimated $K_{ii}$')
plt.xscale('log')
plt.xlabel('Number of Points Sampled (N)')
plt.ylabel('Kinetic Energy Matrix Element ($K_{ii}$)')
plt.title('Kinetic Energy Matrix Element ($K_{ii}$) integration\nRandom Sampling')
plt.grid()
plt.legend()

# The results indicate converge to kii=0.497 with an accurracy to two significant figures. With a random sampling, the accurracy depends on a starting point,
# in this case it started to converge with a low standard deviation of 0.00### with 1x10^7 points, which brings a high computational cost. 
```
