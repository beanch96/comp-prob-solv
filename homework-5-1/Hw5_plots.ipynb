```python
import int_overlap as fn
import matplotlib.pyplot as plt

N_list=[10,100,1000,10000,100000,1000000,10000000,100000000]
R=2   #Separation distance

averages=[]
std_devs=[]

#Go over all the points
for N in N_list:
    integral,std_dev=fn.random_overlap_pz(N,R)
    averages.append(integral)
    std_devs.append(std_dev)
    print(f"N = {N}, S(R) = {integral}")

#Plot Figure
plt.figure(figsize=(8, 6))
plt.errorbar(N_list, averages, fmt='o-',label="S(R)")
plt.xscale('log')
plt.xlabel('Number of Points Sampled')
plt.ylabel('Integral Value')
plt.title('Overlap Integral of Two Pz Orbitals (R=2)\nRandom Sampling')
plt.grid()
plt.legend()
plt.show()

#We can see in the plot that the result of the integration converges to around 0.7362.
#We can be sure of a result up to two significant figures with 1x10^6 points.
#Results from random sampling depends on starting point. Randomly, some starting positions are better than others. 
```

```python
import numpy as np
import int_overlap as fn
import matplotlib.pyplot as plt

R_list = np.arange(0, 25, 0.5)
N=1*10**6

averages=[]


for R in R_list:
    integral,std_dev=fn.important_overlap_pz(N,R)
    averages.append(integral)

plt.figure(figsize=(8, 6))
plt.errorbar(R_list, averages, fmt='o-', label='S(R)')
plt.axhline(0,color="red",linestyle="--")  # =0 overlap when pz orbitals are too far apart
plt.axhline(1,color="red",linestyle="--")  # =1 overlap when pz orbitals have no separation (normalization)
plt.xlabel('Separation Distance (R)')
plt.ylabel('Overlap Integral Value')
plt.title('Overlap Integral of Two Pz Orbitals vs R')
plt.legend()
plt.show()

#At R=0, there is no sepration and the integration essentialy become a normalization test, which as expected becomes 1.
#At R=infinity, there will not be overlap, and as expected it approaches 0. 

#At small R, the two pz orbitals are overlapping in-phase, therefore there is constructive overlap resulting in positive valuues. 
    #With increasing R, the two pz orbitals move apart, and the overlap becomes less significant until it turns negative as the positive overlap
    #of one p orbital starts overlapping with the negative phase in destructive interference.

#The R value when overlap is zero indicates complete cancellation of both phases. 
```
